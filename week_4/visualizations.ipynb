- What is the distribution of the `category_name` column?

```python
 # Subplots makes a call in the background to create a plot area
cat_cnt=df_kick.groupby('category_name',as_index=False).count().sort_values('id',ascending=False)[['category_name','id']]
cat_cnt['Category %']=(cat_cnt['id'] / 
                  cat_cnt['id'].sum()) * 100
plt.figure(figsize=(20,9))
plt.bar(cat_cnt['category_name'].head(10),cat_cnt['Category %'].head(10))
plt.xlabel('Category Name')
plt.ylabel('Percent of Category')
plt.show()
f"I decided to show only the top 10 since they make up the majority of the data"
```

- Which country has the most projects?

```python
sum_cnt=df_kick.groupby('country',as_index=False).count().sort_values('id',ascending=False)[['country','id']].head(5)
plt.figure(figsize=(20,9))
plt.bar(sum_cnt['country'],sum_cnt['id'])
plt.xlabel('Country of Origin')
plt.ylabel('Count of Project Id')
plt.show()

f'The U.S. has the most projects on kickstarter'
```

- What are the top 20 Kickstarter projects?

```python
sum_dlr=df_kick[['name','backers_count']].sort_values('backers_count',ascending=False).head(20)
plt.figure(figsize=(15,12))

plt.bar(sum_dlr['name'],sum_dlr['backers_count'])
plt.xlabel('Project Name')
plt.ylabel('Backer Count')
plt.xticks(rotation=90)
plt.rc('xtick', labelsize=15) 

plt.show()

f'Exploding Kittens is the most backed project, and also it is one of my favorites!'
```

- Which parent category has the most funded projects?

```python
sum_fnd= df_kick.groupby('category_parent_name',as_index=False).sum().sort_values('pledged',ascending=False)[['category_parent_name','pledged']].head(5)
sum_fnd['Millions Pledged']= round((sum_fnd['pledged']/1000000),0)
plt.bar(sum_fnd['category_parent_name'],sum_fnd['Millions Pledged'])
plt.xlabel('Parent Category')
plt.ylabel('Pledged in Funding(in Millions)')
plt.rc('xtick', labelsize=10) 

plt.show()
```
